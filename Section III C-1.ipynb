```python
import numpy as np 

import matplotlib.pyplot as plt


# This import is needed to modify the way figure behaves
from mpl_toolkits.mplot3d import Axes3D
Axes3D
%matplotlib inline 

from ipywidgets import *

import pickle_functions as PK
import plot_functions as PL
import helpers as HL

%load_ext autoreload
import pickle
%autoreload 2
```

```python
folder="SectionC"
```

```python
noises=[0.05,0.1,1,3]
colors=[]
Xs=[]
X_2ds=[]
for i, noise in enumerate(noises):
    color,X,X_2d=PK.get_swiss_roll("lle",folder,modification='_noise'+str(i), create=False, n=1000, noise=noise)
    colors.append(color)
    Xs.append(X)
    X_2ds.append(X_2d)
```

```python
fig = plt.figure(figsize=(10,10))
plot=[221, 222,223,224]

for i in range(len(Xs)):
    X=Xs[i]
    ax = fig.add_subplot(220+i+1, projection='3d')
    ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=colors[i], cmap=plt.cm.Spectral)
    ax.set_title('Noise %1.2f' %noises[i])
plt.show()
```
